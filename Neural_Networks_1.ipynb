```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
# Dense: to add layers (hidden,output)
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
# sequential: to add layers in sequence, to initialize ann model i.e, initially random weight will be assigned
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("/content/Churn_Modelling.csv")
df
# exited is a target column. binary classification problem
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

## **Data Transformation**

```python
df.isna().sum( )
```

```python
df
```

```python
df=pd.get_dummies(data=df,columns=['Geography','Gender'])
df
```

```python
sc=StandardScaler()
x=df.drop(columns=['Exited'])
y=df['Exited']
```

```python
x=sc.fit_transform(x)
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
x_train
```

```python
# create instance of sequential class
model=Sequential()

# Sequential is a class in kearas that allows you to build a model layer-by-layer
# create neurons in input, hidden and output layers and assign random weights to input
# in i/p layer 13 neurons will be there as there are 13 independent features

# Add hidden layer, randomly 10 neurons are added
model.add(Dense(units=10,activation='relu'))# as target is binary use relu activation
#Layers are added sequentially using the .add() method
# Each layer recieves input from the previous layer and passes output to the next one.

# add output layer
model.add(Dense(units=1,activation='sigmoid'))# for binary classification use sigmoid activation in output

# establish the connection b/w the layers
model.compile(optimizer='adadelta',loss='binary_crossentropy',metrics=['accuracy'])
# adadelta is gradient descent algorithm for weight updation which we have selected randomly.we will hypertune it afterly
# binary_crossentropy: also called log_loss, finds errors,range is 0 to 1

#fit the data, perform forward and back propagation
model.fit(x_train,y_train,epochs=100)
#epocs: 100 means that the model will go through the entire x_train and y_train dataset 100 times during training.
#epoch: backward and forward propagation will done for 100 times
# the no of batches per epoch=Total no of samples/Batch size(default is 32)
# 80% of training data=8000 samples.so batches=8000/32=250
# each batch is proceeded in sequence during each epoch
# and after all 250 batches are proceded,one epoch is complete
```

```python
x_test
```

```python
ypred=model.predict(x_test)
ypred # in dataset target column is vlues are 0 and 1, but here output is a continuous number
# because in output layer activation function is sigmoid. it gives u o/p in tems of probability
# in sigmoid if probability is >0.5 it gives u 1 and if probability <0 it should give 0
```

```python
y_test
```

```python
ypred>0.5
```

```python
ypred=ypred>0.5
ypred
```

```python
print(classification_report(y_test,ypred))
```

### **Tuning the ANN model**

```python
import keras_tuner as kt
```

```python
# tuning for best optimizer(adam,rmsprop or sgd:stochastic gradient descent)only
def optimizer_selection(hp):
  # create instence of a sequential class
  model=Sequential()
  #add hidden layer
  model.add(Dense(units=10,activation='relu'))
  # add output layer
  model.add(Dense(units=1,activation='sigmoid'))
  # optimizer selection
  optim=hp.Choice('optimizer',values=['sgd','adam','rmsprop'])# value of optimizer is categorical so use choice() function
  model.compile(optimizer=optim,loss='binary_crossentropy',metrics=['accuracy'])
  return model
```

```python
# similar to GridSearchCV() we will use RandomSearch()
tuner=kt.RandomSearch(
    optimizer_selection,# model name-here sequential() class is our model
    objective='val_accuracy',# increase accuracy of test data i.e, validation testv accuracy or minimize 'val_loss'
    max_trials=3 # for each optimizer take 3 trails, can select 3 to 10 or 20 for larger problems
)
```

```python
tuner.search(x_train,y_train,epochs=3,validation_data=(x_test,y_test))
# tuner.search()is equivalent to grid.fit()
# epochs should be same as max_trails so 3 is selected
#validation_data: pass test data
```

```python
tuner.get_best_hyperparameters()[0].values # asper performence all optimizers are stored in an array
# we want high accuracy optimizer present at index 0 location
```

```python
model=tuner.get_best_models(num_models=1)[0]# get best model: adam
model.fit(x_train,y_train,epochs=100,validation_data=(x_test,y_test))
```

```python
model.evaluate(x_test,y_test)
```

# **Regression**

```python
df=pd.read_csv("/content/Price.csv")
df
```

```python
x=df.iloc[:,1:]
y=df['price']
```

```python
x=sc.fit_transform(x)
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
ann=Sequential()

ann.add(Dense(units=30,activation='relu'))# hidden layer1
ann.add(Dense(units=20,activation='relu'))# hidden layer2

ann.add(Dense(units=1))# output layer
# in regression task no activation function is required in o/p layer as we don't have to convert price to 0 or 1

ann.compile(optimizer='adam',loss='mse')

ann.fit(x_train,y_train,epochs=100,validation_data=(x_test,y_test))
```

```python
yp=ann.predict(x_test)
yp
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,yp)
```
